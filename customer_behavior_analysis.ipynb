```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df.columns
```

```python
labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df.head()
```

```python
# 1. Force the column to lowercase so we don't have to worry about capitals
df['frequency_of_purchases'] = df['frequency_of_purchases'].str.lower()

# 2. Use a lowercase dictionary (easier to type correctly!)
clean_mapping = {
    'fortnightly': 14,
    'weekly': 7,
    'annually': 365,
    'quarterly': 90,
    'bi-weekly': 14,
    'monthly': 30,
    'every 3 months': 90
}

# 3. Apply the mapping
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(clean_mapping)

# 4. Verification
print(f"Missing values now: {df['purchase_frequency_days'].isna().sum()}")
```
